# Import libraries

```python
import pandas as pd
```

# Create a dataset

```python
house_prices = {

    'squareMeters': [154.5, 90, 254.2, 112],
    'hasGarage': ['Yes', 'No', 'No', 'Yes'],
    'price_eur': [150000, 104000, 223400, 92700]
}

house_prices_df = pd.DataFrame(house_prices)
house_prices_df.head()
```

# Typical problem

```python
house_prices_df.to_csv('./house_prices.csv')
```

```python
pd.read_csv('./house_prices.csv').head()
```

# Solution 1 - use **`index_col=0`** parameter of **`.read_csv()`** method!

```python
# Save the dataset
house_prices_df.to_csv('./house_prices.csv', index=True)  # <-- index=True - save also index

# Load the dataset
solution_df = pd.read_csv('./house_prices.csv', index_col=0)  # <-- index_col=0

# Look the result
solution_df.head()
```

# Solution 2 - use **`index=False`** in the **`.to_csv()`** method!

```python
house_prices_df.to_csv('./house_prices_no_index.csv', index=False)  # <-- index=False - no saving index
pd.read_csv('./house_prices_no_index.csv').head()
```
